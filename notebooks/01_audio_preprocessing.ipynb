```python
import os

raw_path = "/content/drive/MyDrive/EmergencyCallProject/dataset/raw_audio"
processed_path = "/content/drive/MyDrive/EmergencyCallProject/dataset/processed_audio"

os.makedirs(processed_path, exist_ok=True)
```

```python
from pydub import AudioSegment
from tqdm import tqdm

files = os.listdir(raw_path)

for file in tqdm(files[:10]):  # first 10 files
    if file.endswith(".mp3"):

        mp3_path = os.path.join(raw_path, file)
        wav_path = os.path.join(processed_path, file.replace(".mp3", ".wav"))

        audio = AudioSegment.from_file(mp3_path)

        # Convert to mono + 16kHz
        audio = audio.set_channels(1)
        audio = audio.set_frame_rate(16000)

        audio.export(wav_path, format="wav")
```

```python
for file in tqdm(files):
    if file.endswith(".mp3"):
        mp3_path = os.path.join(raw_path, file)
        wav_path = os.path.join(processed_path, file.replace(".mp3", ".wav"))

        try:
            audio = AudioSegment.from_file(mp3_path)
            audio = audio.set_channels(1)
            audio = audio.set_frame_rate(16000)
            audio.export(wav_path, format="wav")
        except Exception as e:
            print(f"Error processing {file}: {e}")
```
